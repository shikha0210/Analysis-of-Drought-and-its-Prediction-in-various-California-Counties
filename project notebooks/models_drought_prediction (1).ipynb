```python
import xgboost as xgb

dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)
```

```python
param = {
    'max_depth': 5,  # the maximum depth of each tree
    'eta': 0.5,  # the training step for each iteration
    'silent': 1,  # logging mode - quiet
    'objective': 'multi:softprob',  # error evaluation for multiclass training
    'num_class': 2}  # the number of classes that exist in this datset
num_round = 10000  # the number of training iterations #with 2000 i got 0.7698443421221675 with 10000 got 0.7910409714072661
#with 10000 and eta 0.5 got 0.7927193940032442

```

```python
bst = xgb.train(param, dtrain, num_round)
```

```python
bst.dump_model('dump.raw.txt')
```

```python
preds = bst.predict(dtest)
```

```python
import numpy as np
best_preds = np.asarray([np.argmax(line) for line in preds])
```

```python
from sklearn.metrics import precision_score

print (precision_score(y_test, best_preds, average='macro'))
# >> 1.0
```
